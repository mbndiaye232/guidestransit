### 5. Test de Requête Directe sur l'API Locale
*(Assurez-vous d'avoir démarré votre serveur avec `npm start` dans le dossier backend avant d'exécuter cette cellule)*

```python
import urllib.request
import json

url = "http://localhost:5001/api/assistant/chat"
data = json.dumps({"message": "test"}).encode("utf-8")
req = urllib.request.Request(url, data=data, headers={"Content-Type": "application/json"})

print("--- Requête de test vers le serveur API RAG ---")
try:
    with urllib.request.urlopen(req) as response:
        print(f"Status Code: {response.status}")
        print(f"Response Body:\n{response.read().decode('utf-8')}")
except Exception as e:
    print(f"Erreur rencontrée : {e}")
    if hasattr(e, 'read'):
        print(f"Contenu de la réponse d'erreur :\n{e.read().decode('utf-8')}")
```
